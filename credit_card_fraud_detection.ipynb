```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data['Class'].value_counts()
```

```python
#sepearating data into training and testing datasets
real=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(real.shape)
print(fraud.shape)
```

```python
#statistical measures of real transaction data
real.Amount.describe()
```

```python
#statistical measures of fraud transaction data
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
real_sample=real.sample(n=492)
```

```python
#concatatinating two data frames
new_ds=pd.concat([real_sample,fraud], axis=0)
```

```python
new_ds.head()
```

```python
new_ds.tail()
```

```python
new_ds['Class'].value_counts()
```

```python
new_ds.groupby('Class').mean()
```

```python
#splitting data into features and targets
X=new_ds.drop(columns='Class', axis=1)
Y=new_ds['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#split data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

Training the model

```python
model=LogisticRegression(solver='saga' , max_iter=2000)
```

```python
scaler= StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train_scaled, Y_train)
```

```python
#accuracy score on training data
X_train_prediction = model.predict(X_train_scaled)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_scaled = scaler.fit_transform(X_test)
```

```python
X_test_prediction= model.predict(X_test_scaled)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data: ', test_data_accuracy)
```
